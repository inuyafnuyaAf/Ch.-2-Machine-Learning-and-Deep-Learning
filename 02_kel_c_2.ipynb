### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import cluster
from sklearn.metrics import silhouette_score
```

### Read Dataset

```python
data = pd.read_csv('https://raw.githubusercontent.com/Rietaros/kampus_merdeka/main/cluster_s1.csv')
data = data.drop('no', axis = 1)
data.head(10)
```

```python
%matplotlib inline
ax1 = data.plot.scatter(x='x', y='y', c='DarkBlue')
```

### Data Segmentation (Clustering)

```python
# TODO: Lakukan pencarian jumlah cluster terbaik berdasarkan data diatas

silhouetteScores = []
clusterAmount = 20
clusterRange = range(2, clusterAmount + 1)
for i in clusterRange:
  kmeans = cluster.KMeans(n_clusters=i, random_state=42)
  labels = kmeans.fit_predict(data)
  silhouetteScores.append(silhouette_score(data, labels))

bestCluster = silhouetteScores.index(max(silhouetteScores)) + 2

plt.figure(figsize=(10,6))
plt.plot(clusterRange, silhouetteScores, marker='o')
plt.title("Silhouette Scores for Varying Number of Clusters")
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.xticks(clusterRange)
plt.grid(True)
plt.show()

print(f"The best cluster for this dataset is {bestCluster} with Silhouette Score of {max(silhouetteScores):.5f}")
```

```python
# TODO: Lakukan pemodelan dengan KMeans
kmeans = cluster.KMeans(n_clusters=bestCluster, random_state=42)
kmeans.fit(data)
label = kmeans.fit_predict(data)

score = silhouette_score(data, label)
print("silhouette score={:.5f}".format(score))
```

### Visualization: Cluster Result

```python
# TODO: Masukkan label hasil clustering ke dalam DataFrame

result = data.copy()
result['cluster'] = label
result.head()
```

```python
# TODO: Plot data hasil clustering dengan Seaborn
plt.figure(figsize=(10, 6))

sns.scatterplot(x='x', y='y', hue='cluster', data=result, s=50)

plt.title('KMeans Clustering Results')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
